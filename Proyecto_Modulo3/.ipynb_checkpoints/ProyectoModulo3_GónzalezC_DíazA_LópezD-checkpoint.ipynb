# 1.1 Ecuaciones de Lotka-Volterra: Modelo Presa-Depredador

## 1.2 Objetivos

### 1.2.1 Objetivo general

+ **Modelar** mediante una ecuación diferencial la **interacción entre la población de lobos y ciervos**, lo que se conoce como un **Modelo Presa-Depredador**.

### 1.2.2 Objetivos específicos

+ **Resolver las ecuaciones diferenciales de Lotka-Volterra** con ayuda de la librería *scypy.integrate y su función odeint*.  
+ **Conocer la dinámica e interacción de ambas poblaciones**, al variar distintos *parámetros y condiciones iniciales* del modelo.

## 1.3 Modelo que representa el problema

$$ \frac{dx}{dt}=  \alpha x - \beta x y $$

$$ \frac{dy}{dt}= -\gamma y + \delta x y $$

$\alpha$: tasa de crecimiento de las presas 

$\beta$: éxito en la caza del depredador

$\gamma$: tasa de decrecimiento de los depredadores 

$\delta$: éxito en la caza y cuánto alimenta cazar una presa al depredador

### ¿Qué situación representa el modelo? 

+ La población de **presas** *en ausencia de depredadores* **crece de manera exponencial**.
+ La población de **depredadores** *en ausencia de presas* **decrece de manera exponencial**.
+ *La población de depredadores afecta a la de presas haciéndola decrecer de forma proporcional al número de presas y depredadores* (esto es como decir de forma proporcional al número de posibles encuentros entre presa y depredador).
+ *La población de presas afecta a la de depredadores también de manera proporcional al número de encuentros, pero con distinta constante de proporcionalidad* (dependerá de cuanto sacien su hambre los depredadores al encontrar una presa).


### ¿Cuáles son las limitaciones principales?

+ **El ecosistema está aislado**, es decir *no hay otras especies presentes ni se toman a consideración otros factores (como la migración)*. 
+ Las presas sólo mueren cuando son **cazadas por el depredador**.


## 1.4 Solución del problema

### Condiciones Iniciales

1. $x(0)=\left[x_1(0)\quad x_2(0)\right]^T = [50 \quad 10]^T$.
2. $x(0)=\left[x_1(0)\quad x_2(0)\right]^T = [10 \quad 50]^T$.
3. $x(0)=\left[x_1(0)\quad x_2(0)\right]^T = [50 \quad 50]^T$.
4. $x(0)=\left[x_1(0)\quad x_2(0)\right]^T = [1 \quad 20]^T$.
5. $x(0)=\left[x_1(0)\quad x_2(0)\right]^T = [20 \quad 1]^T$.

### Parámetros

$$ \alpha=0.1  \beta=0.02   \gamma=0.3  \delta=0.01 $$ 

#### Mayor tasa de crecimiento de las presas
$$\alpha=0.8$$   

#### Menor tasa de éxito en la caza
$$\beta=0.001$$ 

#### Menor tasa de crecimiento de las presas y mayor tasa de decrecimiento de los depredadores
$$\alpha=0.001$$  
$$\gamma=0.5$$ 

#### Mayor tasa de crecimiento de las presas y mayor tasa de alimentación a los depredadores
$$\alpha=0.3$$
$$\gamma=0.5$$
$$\delta=0.2$$ 

```python
# Importar librerías
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from scipy.integrate import odeint
```

```python
def presa_depredador(x, t):
    x1 = x[0]   #Presas
    x2 = x[1]   #Depredadores
    a=0.1
    b=0.02
    c=0.3
    d=0.01
    dx1 = (a*x1)-(b*x1*x2)
    dx2 = (-c*x2)+(d*x1*x2)
    return np.array([dx1, dx2])
```

```python
# Primer condición inicial
x0 = np.array([50, 10])    
# Vector de tiempo
t = np.linspace(0, 200, 800)
```

```python
# Resolver numéricamente
xx = odeint(presa_depredador, x0, t)
# Separamos soluciones
x1 = xx[:, 0]
x2 = xx[:, 1]
```

## 1.5 Visualización de la solución del problema

```python
# Gráficas de x1 vs. t y x2 vs. t
plt.figure(figsize=(8,6))
plt.plot(t, x1, label='Presa: Ciervo $x_1(t)$')
plt.plot(t, x2, label='Depredador: Lobo $x_2(t)$')
plt.xlabel('Tiempo $t$')
plt.ylabel("Población")
plt.title("Modelo Presa-Depredador")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(x1, x2, label='Trayectoria de poblaciones')
plt.plot(x1[0], x2[0], '*', ms=10, label = 'Condiciones iniciales')
plt.plot(x1[-1], x2[-1], '*', ms=10, label = 'Condiciones finales')
plt.xlabel('Población de ciervos $x_1$')
plt.ylabel('Población de lobos $x_2$')
plt.title("Modelo Presa-Depredador")
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.legend(loc = 'best')
plt.grid()
plt.show()
```
